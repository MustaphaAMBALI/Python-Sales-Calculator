# Phase One Deliverables

```python
import csv

valid_rows = []
skipped_rows = []  

print("── Sales Calculator ──────────────────")

with open(r"phase_one.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        try:
            order_id   = row["order_id"]
            rep_name   = row["rep_name"]
            category   = row["category"]
            product    = row["product"]
            
            if not row["units_sold"]:
                raise ValueError("missing units_sold")
            if not row["unit_price"]:
                raise ValueError("missing unit_price")

            if row["units_sold"].isdigit() == False:
                raise ValueError(f"invalid units_sold value: '{row['units_sold']}'")

            units_sold = int(row["units_sold"])
            unit_price = int(row["unit_price"])
            
            if not category:  # ← catches empty string
                raise ValueError("missing category")
            
            revenue = units_sold * unit_price
            valid_rows.append({
                "order_id":  order_id,
                "rep_name":  rep_name,
                "category":  category,
                "product":   product,
                "revenue":   revenue
            })
        except ValueError as e:
            skipped_rows.append({
                "order_id": row["order_id"],
                "rep_name": row["rep_name"],
                "reason": str(e)  # ← captures the error message as the reason
            })

print(f"Valid rows processed: {len(valid_rows)}")
print(f"Rows skipped: {len(skipped_rows)}")

category_revenue = {}

for row in valid_rows:
    name = row["category"]
    revenue = row["revenue"]

    if name in category_revenue:
        category_revenue[name] += revenue
    else:
        category_revenue[name] = revenue
        
print(f"\n── Revenue by Category ───────────────")
for category, total in category_revenue.items():
    print(f"{category:<15} ${total:,}")

print(f"\n── Overall Summary ───────────────")
print(f"Total Revenue: ${sum(row["revenue"] for row in valid_rows):,}")

top_category = max(category_revenue, key=lambda cat: category_revenue[cat])
print(f"Top Category: {top_category} - ${category_revenue[top_category]:,}")

total_revenue = sum(row["revenue"] for row in valid_rows)
average_order = total_revenue / len(valid_rows)
print(f"Average Order: ${average_order:,.0f}")
 
print(f"\n── Skipped Rows ──────────────────────")
for row in skipped_rows:
    print(f"Row {row['order_id']} - {row['rep_name']}: {row['reason']}")

with open("sales_summary.txt", "w", encoding="utf-8") as report:
    report.write("── Sales Calculator Report ──────────────────\n")
    report.write(f"Valid rows processed: {len(valid_rows)}\n")
    report.write(f"Rows skipped: {len(skipped_rows)}\n")
    report.write(f"Total Revenue: ${sum(row['revenue'] for row in valid_rows):,}\n")
    report.write(f"Top Category: {top_category} - ${category_revenue[top_category]:,}\n")
    report.write(f"Average Order: ${average_order:,.0f}\n")
    report.write("\n── Skipped Rows ──────────────────────\n")
    for row in skipped_rows:
        report.write(f"Row {row['order_id']} - {row['rep_name']}: {row['reason']}\n")

print("\n📄 Report saved to sales_cli_report.txt")
```
